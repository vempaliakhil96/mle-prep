# Product Sense Case Study Frameworks

## Diagnose a problem

Typically you would be given a metric / KPI that is not performing as expected. You would then need to diagnose the problem. 

- Clarify the metric / KPI
    - What does the metric measure?
    - How is it measured?
    - What is the frequency of measurement?
- Temporal analysis
    - How has the metric been performing over time?
        - week-on-week
        - year-on-year
    - Are there any trends?
        - industry trend?
        - seasonality?
    - external factors?
        - change by competition
        - drastic change in environment
- Product related or company related change
    - any change in product or feature?
    - any change in company policy?
    - change in competitors product?
- Demographic / geographic / behavioral analysis
    - isolate based on region / location
    - segment by language
    - platforms
        - android vs ios vs web
        - mobile vs desktop
- Decompose metric
    - is metric a composite of other metrics?
    - is the mteric predicted?    

## Measure Success of product / feature

- clarify the function or goal of the product / feature
    - what does the product do?
    - How is it used?
    - Who is it for?
- define metric
    - success metric
    - guardrail metric
    - should fit context

## Launch or not

- clarify the goal and define metric
    - ask question to clarify goal
    - define a success metric
    - define a guardrail metric
- experimentation
    - how to design the A/B test
    - how to split the users
    - How long to run it?
- recommendation
    - conflicting results
