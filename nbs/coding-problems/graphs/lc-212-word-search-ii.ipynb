# 212. Word Search II

## Problem

[link](https://leetcode.com/problems/word-search-ii/)

Given an `m x n `board of characters and a list of strings `words`, return all words on the board.

Each word must be constructed from letters of sequentially **adjacent cells**, where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once in a word.

![example 1](../../assets/lc-212-word-search-ii-1.jpeg)
```code
Input: board = [["o","a","a","n"],["e","t","a","e"],["i","h","k","r"],["i","f","l","v"]], words = ["oath","pea","eat","rain"]
Output: ["eat","oath"]
```

![example 2](../../assets/lc-212-word-search-ii-2.jpeg)
```code
Input: board = [["a","b"],["c","d"]], words = ["abcb"]
Output: []
```

## Constraints

- `m == board.length`
- `n == board[i].length`
- `1 <= m, n <= 12`
- `board[i][j]` is a lowercase English letter.
- `1 <= words.length <= 3 * 10^4`
- `1 <= words[i].length <= 10`
- `words[i]` consists of lowercase English letters.
= All the strings of words are unique.

````
from collections import defaultdict
from typing import List
from functools import lru_cache

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:

        class TrieNode:
            def __init__(self):
                self.is_word = False
                self.children = defaultdict(TrieNode)
            
            def insert(self, word):
                n = self
                for c in word: n = n.children[c]
                n.is_word = True
                return True
            
            @lru_cache
            def search(self, word):
                n = self
                for c in word:
                    if c not in n.children.keys(): return (False, False)
                    n = n.children[c]
                return (True, n.is_word)
            
            def remove(self, word):
                n = self
                for c in word:
                    if c not in n.children.keys(): return False
                    n = n.children[c]
                n._is_word = False
                return True

        result, root = [], TrieNode()
        [root.insert(w) for w in words] # insert all words into trie
        
        def dfs(i, j, node, ws=""):
            if node.is_word: # if word is found, add to result and remove it from trie
                result.append(ws); node.is_word = False
            if not(0<=i<len(board)) or not(0<=j<len(board[0])): return

            c = board[i][j]
            node = node.children.get(c, False) # get next node if present else return
            if not node: return
            board[i][j] = "#" # mark visited
            for ih, jh in [(i+1,j), (i-1,j), (i, j+1), (i,j-1)]:
                dfs(ih, jh, node, ws+c)
            board[i][j] = c # backtrack
            return
        
        for i in range(len(board)):
            for j in range(len(board[0])):
                dfs(i, j, root)
        
        return list(set(result))
````
