# 1584. Min Cost to Connect All Points

## Problem

[link](https://leetcode.com/problems/min-cost-to-connect-all-points/)

You are given an array `points` representing integer coordinates of some points on a 2D-plane, where `points[i] = [xi, yi]`.

The cost of connecting two points `[xi, yi]` and `[xj, yj]` is the **manhattan distance** between them: `|xi - xj| + |yi - yj|`, where `|val|` denotes the absolute value of `val`.

Return the *minimum* cost to make all points connected. All points are connected if there is **exactly one** simple path between any two points.

![](../../assets/lc-1584-ex1.png)
```code
Input: points = [[0,0],[2,2],[3,10],[5,2],[7,0]]
Output: 20
Explanation:
```
![](../../assets/lc-1584-ex1-1.png)
```
We can connect the points as shown above to get the minimum cost of 20.
Notice that there is a unique path between every pair of points.
Example 2:

Input: points = [[3,12],[-2,5],[-4,1]]
Output: 18
```
## Constraints

- `1 <= points.length <= 1000`
- `-10^6 <= xi, yi <= 10^6`
- All pairs `(xi, yi)` are distinct.

## Solution

````
import heapq
from typing import List

class DisJointSetUnion:
    def __init__(self, n):
        self.parents = list(range(n))
    
    def find(self, x):
        if self.parents[x]==x: return x
        r = self.find(self.parents[x]); return r
    
    def union(self, x, y): self.parents[self.find(y)] = self.find(x)

class Solution:
    def minCostConnectPoints(self, points: List[List[int]]) -> int:

        n = len(points)
        UF = DisJointSetUnion(n)
        distances = []

        for i in range(n):
            for j in range(i+1, n):
                x1, y1 = points[i]
                x2, y2 = points[j]
                d = abs(x1-x2)+abs(y1-y2)
                heapq.heappush(distances, (d, i, j))
        
        count, total_cost = 0, 0
        while distances:
            d, i, j = heapq.heappop(distances)
            if UF.find(i)==UF.find(j) or count==n-1: continue
            UF.union(i, j)
            total_cost += d
            count += 1
        
        return total_cost
````
